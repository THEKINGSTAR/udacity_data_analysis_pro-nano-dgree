```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("winequality_edited.csv")
df.head(2)
```

```python
df.groupby('color')['quality'].mean().plot(kind='bar',title='Average wine by Color',color=['red','white'],alpha=.7);
```

```python
colors = ['red','white']
df.groupby([df.color,df.quality]).mean().plot(kind='bar',title='Average wine by Color',color=colors,alpha=.7);
```

```python
colors = ['red','white']
mean= df.groupby('color')['quality'].mean().plot(kind='bar',title='Average wine by Color',color=colors,alpha=0.5);
plt.xlabel('Colors',fontsize=18)
plt.ylabel('Quality',fontsize=18)
```

```python
colors = ['red','green']
df_mean = df.groupby('color')['quality'].mean()
df_mean.plot(kind='bar',title='Average wine by Color',color=colors,alpha=0.5);
plt.xlabel('Colors',fontsize=18)
plt.ylabel('Quality',fontsize=18)
```

```python
counts = df.groupby(['quality','color']).count()["pH"]
counts
```

```python
counts.plot(kind='bar',title="Count by Wins color and Quality",color=colors,alpha=.7)
plt.xlabel("Qulity and Color", fontsize=18)
plt.ylabel("Count and Color" , fontsize=18)
```

```python
total = df.groupby('color').count()["pH"]
proportions = counts/total

proportions.plot(kind='bar',title="Count by Wins color and Quality",color=colors,alpha=.7)
plt.xlabel("Quality and Color",fontsize=19)
plt.ylabel("Proportion",fontsize=18)
```
